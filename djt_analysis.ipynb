# Sample sentiment analysis based on sentence


```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

#sample sentiment analysis
analyzer = SentimentIntensityAnalyzer()

sentence = "I love PyCharm! It's my favorite Python IDE."
sentiment_scores = analyzer.polarity_scores(sentence)
print(sentiment_scores)
```

# Sentiment Analysis for a random sample of 50 tweets
## Added negative/positive polarity score to new column

```python
#take sample of 50 rows from dataframe
sample_df = df.sample(n=50, random_state=42)

#apply sentiment analysis on sample dataframe and create new columns for negative and positive scores
sample_df['negative_score'] = sample_df['text'].apply(lambda x: analyzer.polarity_scores(x)['neg'] if pd.notnull(x) else None)
sample_df['positive_score'] = sample_df['text'].apply(lambda x: analyzer.polarity_scores(x)['pos'] if pd.notnull(x) else None)
print(sample_df[['text', 'positive_score']])
```
